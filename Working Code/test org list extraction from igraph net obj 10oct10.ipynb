```r
nodes <- read.csv(file.choose())
```

```r
links <- read.csv(file.choose()
                 )
```

```r
nodes <- nodes[,c(1:9)]
nodes <- unique(nodes)
net <- graph.data.frame(links, directed = F, vertices = nodes)
```

```r
plot(net)
```

```r
V(net)$degree <- degree(net)
V(net)$degree <- V(net)$degree / ( nrow(nodes) - 1 )
vertex.attributes(net)
```

```r
nodes <- data.frame(matrix(NA,
                             ncol = length(vertex.attributes(net)),
                             nrow = vcount(net)
                             )
                      )
print(names(vertex.attributes(net)))
for (attr in 1:length(vertex.attributes(net))) {
    print(attr)
    nodes[,attr] <- data.frame(vertex.attributes(net)[attr])
    }
colnames(nodes) <- names(vertex.attributes(net))
nodes
```
